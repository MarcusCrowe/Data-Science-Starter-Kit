## Descriptive Statistics

```python
# TODO: Load the dataset and compute mean, median, mode, std, IQR
import pandas as pd
import numpy as np
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv')
victims = df['Nï¿½mero de vï¿½ctimes']
print('Mean:', victims.mean())
print('Median:', victims.median())
print('Mode:', victims.mode().iloc[0])  # mode() returns a Series
print('Standard Deviation:', victims.std())
print('Interquartile Range (IQR):', victims.quantile(0.75) - victims.quantile(0.25))
```

## Visualizations (Histogram, KDE, Boxplot, Heatmap)

```python
# TODO: Plot histogram, KDE, boxplot, and correlation heatmap
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data with proper encoding
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv', encoding='latin1')

victims = df['NÃ¯Â¿Â½mero de vÃ¯Â¿Â½ctimes'].dropna()

# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(victims, bins=15, kde=False, color='skyblue', edgecolor='black')
plt.title('Histogram of Number of Victims')
plt.xlabel('Number of Victims')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(victims, fill=True, color='purple')
plt.title('Kernel Density Estimate (KDE) of Victims')
plt.xlabel('Number of Victims')
plt.ylabel('Density')
plt.tight_layout()
plt.show()

# Boxplot
plt.figure(figsize=(8, 2))
sns.boxplot(x=victims, color='lightgreen')
plt.title('Boxplot of Number of Victims')
plt.xlabel('Number of Victims')
plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap (Numeric Columns)')
plt.tight_layout()
plt.show()
```

## Bootstrap Confidence Interval

```python
# TODO: Bootstrap sample mean and CI (e.g., for 'NÃºmero de vÃ­ctimes')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv', encoding='latin1')

victims = df['NÃ¯Â¿Â½mero de vÃ¯Â¿Â½ctimes'].dropna()

# Bootstrap sampling
boot_means = [victims.sample(frac=1, replace=True).mean() for _ in range(1000)]

# Calculate 95% confidence interval
ci_lower = np.percentile(boot_means, 2.5)
ci_upper = np.percentile(boot_means, 97.5)

print(f'95% CI for mean victims: ({ci_lower:.2f}, {ci_upper:.2f})')

# Optional: Plot the bootstrap distribution
plt.figure(figsize=(8, 5))
sns.histplot(boot_means, bins=30, kde=True, color='lightcoral')
plt.axvline(ci_lower, color='blue', linestyle='--', label=f'2.5% = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'97.5% = {ci_upper:.2f}')
plt.title('Bootstrap Distribution of Mean Victims')
plt.xlabel('Mean Number of Victims')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()

```

## Hypothesis Testing

```python
# TODO: Perform a t-test (e.g., is mean != 1?) and interpret the result
from scipy.stats import ttest_1samp
import pandas as pd

# Load your data
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv', encoding='latin1')

victims = df['NÃ¯Â¿Â½mero de vÃ¯Â¿Â½ctimes'].dropna()

# One-sample t-test against population mean = 1
t_stat, p_val = ttest_1samp(victims, popmean=1)

print(f't-statistic: {t_stat:.2f}, p-value: {p_val:.4f}')
if p_val < 0.05:
    print('Reject null hypothesis: mean is significantly different from 1')
else:
    print('Fail to reject null hypothesis')
```
